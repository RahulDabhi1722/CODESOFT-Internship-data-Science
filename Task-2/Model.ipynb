```python
import pandas as pd
#read data from csv file
data = pd.read_csv("IMDb Movies India.csv", encoding="latin1")
print(data)
print(data.info())
```

```python
#print first some data
print(data.head())
```

```python
#print first last data
print(data.tail())
```

```python
data.describe()
```

```python
#check missing value or NULL 
data.isnull().sum()
```

```python
#we have use dropna function on Rating
data.dropna(subset=["Rating"], inplace = True)
```

```python
data.isnull().sum()
```

```python
#Again we have use dropna function 
data.dropna(subset=['Actor 1','Actor 2','Actor 3','Director','Genre'],inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
# convert votes columns to integer
data['Votes'].apply(str).str.replace(',', '')
```

```python
# convert year columns
data['Year'] = data['Year'].str.strip('()').astype(int)
```

```python
# convert duration columns
data['Duration'] =data['Duration'].str.strip('min')
```

```python
data['Duration'].fillna(data['Duration'].median(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.head()
#Now Data is Clean...
```

```python
#Phase of Data Analysis
# find top 10 movies based on rating
topMovie = data.loc[data['Rating'].sort_values(ascending=False)[:10].index]
topMovie
```

```python
#Distribution of Top 10 movies wrt Year
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(data=topMovie, x="Year",hue="Rating", multiple="stack")
plt.title('Distribution of Top 10 Movies', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Rating', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Distribution of Genre movies
GenreCounts = data['Genre'].value_counts().reset_index()
GenreCounts.columns = ['Genre', 'Count']

# Select the top N genres (e.g., top 5)
topGenres = GenreCounts.head(5)
topGenres
```

```python
plt.figure(figsize=(4, 4))
plt.pie(topGenres['Count'], labels=topGenres['Genre'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Movie Genre Distribution', fontsize=16)
plt.axis('equal')
plt.show()
```

```python
# Group the data by director and calculate the average rating
Director_AvgRating = data.groupby('Director')['Rating'].mean().reset_index()

Director_AvgRating = Director_AvgRating.sort_values(by='Rating', ascending=False)

TopDirectors = Director_AvgRating.head(10)
TopDirectors
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=TopDirectors, x='Rating', y='Director', palette='viridis')

plt.title('Top Directors by Average Rating', fontsize=16)
plt.xlabel('Average Rating', fontsize=14)
plt.ylabel('Director', fontsize=14)

plt.show()
```

```python
#Relationship between the number of votes and movie ratings
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Rating', y='Votes')

plt.title('Votes vs. Rating', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Votes', fontsize=14)

plt.show()
```

```python
#Distribution of top actors by number of movie
ActorCounts = data['Actor 1'].value_counts().reset_index()
ActorCounts.columns = ['Actor', 'MovieCount']

TopActors = ActorCounts.head(10)
TopActors
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=TopActors, x='MovieCount', y='Actor', orient='h')

# Set plot labels and title
plt.title('Top Actors by Number of Movies', fontsize=16)
plt.xlabel('Number of Movies', fontsize=14)
plt.ylabel('Actor', fontsize=14)

# Show the plot
plt.show()
```

```python
#Distribution of number of movie released every year
Yearly_MovieCounts = data['Year'].value_counts().reset_index()
Yearly_MovieCounts.columns = ['Year', 'MovieCount']

Yearly_MovieCounts = Yearly_MovieCounts.sort_values(by='Year')
Yearly_MovieCounts
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=Yearly_MovieCounts, x='Year', y='MovieCount')

plt.title('Number of Movies Released Every Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Movies Released', fontsize=14)

plt.show()
```

```python
filtered = data[(data['Rating'] < 2) & (data['Votes'] > 100)]
print(filtered.head(10))
```

```python
plt.figure(figsize=(15, 6))
ax=sns.barplot(data=filtered,x='Name',y='Votes',hue='Rating',dodge=False,width=0.5,palette='muted')

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right')
ax.legend(loc='upper right')
ax.set_xlabel('Movie Name')
ax.set_ylabel('Votes')
ax.set_title('Movies with rating less than 2 and votes greater than 100')

plt.show()
```

```python
#Distribution of Durations of movies by years
import numpy as np
data['Duration'] = data['Duration'].astype(int)
data['Year'] = data['Year'].astype(int)


plt.figure(figsize=(12, 6))
sns.lineplot(data=data,x='Year',y='Duration',errorbar=None)
plt.xlabel('Year')
plt.ylabel('Duration in minutes')
plt.title('Duration of movies by year')
plt.xticks(np.arange(1917,2023,5))
plt.show()
```

```python
#Distribution of Number of movies each genre
data['Genre'] = data['Genre'].str.split(',')

# Create a new DataFrame with one row for each genre
Genre = data.explode('Genre')
Genre
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=Genre, x='Genre', order=Genre['Genre'].value_counts().index, palette='viridis')

plt.title('Number of Movies for Each Genre', fontsize=16)
plt.xlabel('Number of Movies', fontsize=14)
plt.ylabel('Genre', fontsize=14)
plt.xticks(rotation=90)

plt.show()
```

```python
#Distribution of Average rating of movies in each genre
Average_Rating= Genre.groupby('Genre')['Rating'].mean().reset_index()
Average_Rating= Average_Rating.sort_values(by='Rating', ascending=False)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=Average_Rating,y="Rating",x='Genre',palette='coolwarm')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average rating of movies in each genre')
plt.xticks(rotation=90)

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

Trans_Data = data.drop(['Name'],axis=1)
# Transform Director columns
Trans_Data['Director'] = labelencoder.fit_transform(data['Director'])

# Transform Actors Columns
Trans_Data['Actor 1'] = labelencoder.fit_transform(data['Actor 1'])
Trans_Data['Actor 2'] = labelencoder.fit_transform(data['Actor 2'])
Trans_Data['Actor 3'] = labelencoder.fit_transform(data['Actor 3'])

Trans_Data['Genre'] = labelencoder.fit_transform(data['Genre'].apply(lambda x: ', '.join(x)))

Trans_Data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
Sc_Data = scaler.fit_transform(Trans_Data)
Sc = pd.DataFrame(Sc_Data,columns=Trans_Data.columns)

Sc.head()
```

```python
# correlation
Corr = Trans_Data.corr(numeric_only=True)
Corr['Rating'].sort_values(ascending=False)
```

```python
sns.heatmap(Corr,annot=False,cmap="coolwarm")
```

```python
#At A last phase we can construct Model
# Import modeling Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
# put data except Rating data
X = Trans_Data.drop(['Rating'],axis=1)

# Put only Rating data
y = Trans_Data['Rating']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

model = LinearRegression()
model.fit(X_train,y_train)
     
```

```python
X_test = np.array(X_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print('R2 score: ',r2_score(y_test,y_pred))
print('Mean squared error: ',mean_squared_error(y_test,y_pred))
print('Mean absolute error: ',mean_absolute_error(y_test,y_pred))
```
